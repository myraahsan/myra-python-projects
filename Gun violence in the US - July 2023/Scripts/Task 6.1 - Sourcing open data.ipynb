## Creating a dataset df_new 

Creating a separate dataset to understand and check if there are inconsistencies or missing data among some columns

```python
new_df = df[['incident_id', 'participant_age_group', 'participant_gender', 'participant_status', 'participant_type']]
```

```python
new_df.head()
```

Before I make changes to these columns, I need to address the inconsistencies. For example, all the four columns consisting of gender, status, type and age group do not have complete information. In the second row, there are atleast four people involved, which is seen by age groups but in the gender column only one participant is mentioned. 

I need to count the number of participants in each cell and by counting '::', I can derive at the count of that cell

## Creating more columns consisting of counts of age group, gender, status and type

```python
new_df['count_age_group'] = new_df['participant_age_group'].str.count('::')
```

```python
new_df['count_gender'] = new_df['participant_gender'].str.count('::')
new_df['count_status'] = new_df['participant_status'].str.count('::')
new_df['count_type'] = new_df['participant_type'].str.count('::')
```

```python
new_df.head()
```

```python
# Changing the type from floats to integers
new_df[['count_age_group', 'count_gender', 'count_status', 'count_type']] = new_df[['count_age_group', 'count_gender', 'count_status', 'count_type']].fillna(0).astype(int)
```

```python
new_df.head()
```

```python
max_value = new_df[['count_age_group', 'count_gender', 'count_status', 'count_type']].max(axis=1)
```

```python
new_df['diff_age_group'] = max_value - new_df['count_age_group']
new_df['diff_gender'] = max_value - new_df['count_gender']
```

```python
new_df['diff_status'] = max_value - new_df['count_status']
new_df['diff_type'] = max_value - new_df['count_type']
```

```python
new_df.head()
```

In the above data, for example, diff_gender shows the number of unknown genders in the data. I will now address them one by one

### Separating and counting age groups

```python
# The 'diff_age_group' column is actually the count of age group which is unknown so we will rename the column to unknown_gender
new_df = new_df.rename(columns={'diff_age_group': 'unknown_age_group'})
```

In the participant_age_group column three types of age groups are written - Child 0-11, Teen 12-17, Adult 18+

```python
# Counting the age groups and putting them in separate columns for each incident 
new_df['Child 0-11 age'] = new_df['participant_age_group'].str.count('Child 0-11')
```

```python
new_df['Teen 12-17 age'] = new_df['participant_age_group'].str.count('Teen 12-17')
new_df['Adult 18+ age'] = new_df['participant_age_group'].str.count('Adult 18+')
```

```python
new_df.head()
```

```python
# Since the age columns are floats, I will change the datatype to integers
new_df[['Teen 12-17 age', 'Adult 18+ age', 'Child 0-11 age']] = new_df[['Teen 12-17 age', 'Adult 18+ age', 'Child 0-11 age']].fillna(0).astype(int)
```

```python
# Renaming the age groups again as I want to change the format of the way I have written
new_df = new_df.rename(columns={'Teen 12-17 age': 'teen_12-17', 'Adult 18+ age': 'adult_18+', 'Child 0-11 age': 'child_0-11'})
```

### Separating and counting gender

```python
# The 'diff_gender' column is actually the count of gender which is unknown so we will rename the column to unknown_gender
new_df = new_df.rename(columns={'diff_gender': 'unknown_gender'})
```

```python
# Counting Male and Female separately and putting them in columns for each incident 
new_df['male_count'] = new_df['participant_gender'].str.count('Male')
new_df['female_count'] = new_df['participant_gender'].str.count('Female')
```

```python
new_df[['male_count', 'female_count']] = new_df[['male_count', 'female_count']].fillna(0).astype(int)
```

```python
new_df.head()
```

### Separating and counting status

```python
# The 'diff_status' column is actually the count of statuses which is unknown so we will rename the column to unknown_status
new_df = new_df.rename(columns={'diff_status': 'unknown_status'})
```

```python
# Counting statuses separately and putting them in columns for each incident 
new_df['arrested'] = new_df['participant_status'].str.count('Arrested')
new_df['injured'] = new_df['participant_status'].str.count('Injured')
new_df['killed'] = new_df['participant_status'].str.count('Killed')
new_df['unharmed'] = new_df['participant_status'].str.count('Unharmed')
```

```python
# Changing the datatype
new_df[['arrested', 'injured', 'killed', 'unharmed']] = new_df[['arrested', 'injured', 'killed', 'unharmed']].fillna(0).astype(int)
```

```python
new_df.head()
```

### Separating and counting participant_type

```python
# The 'diff_type' column is actually the count of type which is unknown so we will rename the column to unknown_type
new_df = new_df.rename(columns={'diff_type': 'unknown_type'})
```

```python
# Counting types separately and putting them in columns for each incident 
new_df['suspect'] = new_df['participant_type'].str.count('Subject-Suspect')
new_df['victim'] = new_df['participant_type'].str.count('Victim')
```

```python
# Changing the datatype
new_df[['suspect', 'victim']] = new_df[['suspect', 'victim']].fillna(0).astype(int)
```

### Deleting extra columns from both dataframes

```python
# Now that I have addressed gender, types, status and age groups, I will delete extra columns that I do not need from the original dataframe before merging it with new_df
new_df = new_df.drop(['participant_age_group', 'participant_gender', 'participant_status', 'participant_type', 'count_age_group', 'count_gender', 'count_status', 'count_type'], axis=1)
```
