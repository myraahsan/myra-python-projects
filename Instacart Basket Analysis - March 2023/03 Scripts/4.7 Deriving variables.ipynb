## Defining If-Statements

```python
# define a function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Checking for the most expensive product
df['prices'].max()
```

```python
# Checking for the value count of 'price_range' column
df.value_counts('price_range')
```

The above data shows that there are no high range products and missing data in the 'prices' column in this given subset

## If-Statement with For-Loops in ords_prods_merged

```python
# To know on which day most orders take place
ords_prods_merged['order_day_of_week'].value_counts(dropna = False)
```

The most orders are on 0.0 which is Saturday and the least are on 4.0, which represents Wednesday (this information is given in the Project Brief)

```python
result = []

for value in ords_prods_merged["order_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

#### Creating 'busiest_day' column

```python
# Creating a new column within your ords_prods_merged dataframe and setting it equal to the result
ords_prods_merged['busiest_day'] = result
```

```python
ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

### Using If-statement for For-Loops in checking busiest days of the week

```python
result_2 = []

for value in ords_prods_merged['order_day_of_week']:
  if (value == 0) or (value == 1):
    result_2.append('Busiest days')
  elif (value == 4) or (value == 3):
    result_2.append('Least busy days')
  else:
    result_2.append('Regularly busy days')
```

#### Creating 'busiest_days_of_week' column

```python
# Creating a new column within your ords_prods_merged dataframe and setting it equal to the result_2
ords_prods_merged['busiest_days_of_week'] = result_2
```

```python
ords_prods_merged['busiest_days_of_week'].value_counts(dropna = False)
```

```python
# Checking the output of the 'order_day_of_week' and the 'busiest_days_of_week' column for accuracy
ords_prods_merged.loc[:,['order_day_of_week', 'busiest_days_of_week'] ].head(15)
```

### Using If-statement for For-Loops in checking busiest hours of the day

```python
ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Divide the hours of the day in three groups, each having eight values
result_3 = []

for value in ords_prods_merged['order_hour_of_day']:
  if (value == 10) or (value == 11) or (value == 14) or (value == 15) or (value == 13) or (value == 12) or (value == 16) or (value == 19):
    result_3.append('Most orders')
  elif (value == 17) or (value == 8) or (value == 18) or (value == 19) or (value == 20) or (value == 7) or (value == 21) or (value == 22):
    result_3.append('Average orders')
  else:
    result_3.append('Fewest orders')
```

```python
result_3
```

#### Creating 'busiest_period_of_day' column

```python
# Creating a new column
ords_prods_merged['busiest_period_of_day'] = result_3
```

```python
# Checking the output of the 'order_hour_of_day' and the 'busiest_days_of_week' column for accuracy
ords_prods_merged.loc[:,['order_hour_of_day', 'busiest_period_of_day'] ].head(15)
```

```python
# Checking the frequency of the values in the 'busiest_period_of_day' column
ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```
