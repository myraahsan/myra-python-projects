## Grouping and aggregating data

```python
# This will group the dataframe by 'product_name' column
df.groupby('product_name')
```

```python
df.groupby('department_id').agg({'order_number': ['mean']})
```

The table above shows the average number of orders in per department. 

```python
# Producing multiple statistics at the same time
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
orders_products_merged['max_order'] = orders_products_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
orders_products_merged.head(15)
```

A new “max_order” column has appeared at the end of the dataframe. Each value in this column corresponds to the maximum number of orders made by each user ID.

```python
# Creating a flag that assigns a “loyalty” label to a user ID based on its corresponding max order value.
orders_products_merged.loc[orders_products_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
orders_products_merged.loc[(orders_products_merged['max_order'] <= 40) & (orders_products_merged['max_order'] > 10) , 'loyalty_flag'] = 'Regular customer'
orders_products_merged.loc[orders_products_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
orders_products_merged['loyalty_flag'].value_counts(dropna = False)
```
